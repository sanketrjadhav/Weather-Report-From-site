# Access the site

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
source = requests.get('https://weather.com/en-IN/weather/tenday/l/7e680b4c959f0c4b864e565a6774292a89e261f5f5db5ae0d9ea39dfd6bc79b3').text
page = BeautifulSoup(source, "lxml")
```

# Inspect the site carefully

```python
data=page.find('tbody')
```

# Define to store the required data

```python
days=[]
dates=[]
weas=[]
descs=[]
temps=[]
winds=[]
hums=[]
```

#  Extract all required data

```python
for body in data.find_all('tr'):
    day = body.find(class_='date-time').text
    days.append(day)
    date = body.find(class_='day-detail clearfix').text
    dates.append(date)
    wea = body.find('td',class_='description').text
    weas.append(wea)
    desc = body.find('td',class_='description')
    desc = desc['title']
    descs.append(desc)
    temp = body.find('td',class_='temp').text
    temps.append(temp)
    wind = body.find('td', class_='wind').text
    winds.append(wind)
    hum = body.find('td',class_='humidity').text
    hums.append(hum)
```

# Store data in DataFrame

```python
weather = pd.DataFrame()
weather["Day"] = days
weather["Date"] = dates
weather["Weather"] = weas
weather["Description"] = descs
weather["Temp"] = temps
weather["Wind"] = winds
weather["Humidity"] = hums
weather
```

# Store the file in csv format

```python
weather.to_csv("Weather_Report.csv", index=False)
```
